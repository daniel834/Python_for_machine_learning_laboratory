```python

import pandas as pd





# CLEANING

def wrangle(filepath):

    # Read CSV file
    df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/heart_disease.csv')

    # Removing null values
    df.dropna(inplace=True)

    # Gender - From object to int - Coding 0 for Male and 1 for Female
    df["gender"] = (df["Gender"].str.replace("Male", "0", regex=False).str.replace("Female", "1", regex=False).astype(int))
    df.drop(columns=["Gender"],inplace=True)

    # education - From object to int - Coding: [(0)'uneducated'], [(1)'primaryschool'], [(2)'graduate'], [(3)'postgraduate']
    df["Education"] = (df["education"].str.replace("uneducated", "0", regex=False)
                   .str.replace("primaryschool", "1", regex=False)
                   .str.replace("graduate", "2", regex=False)
                   .str.replace("postgraduate", "3", regex=False)
                   .str.replace("post2", "3", regex=False)
                   .astype(int))
    df.drop(columns=["education"], inplace=True)

    df["PrevalentStroke"] = (df["prevalentStroke"].str.replace("no", "0", regex=False).str.replace("yes", "1", regex=False).astype(int))
    df.drop(columns=["prevalentStroke"],inplace=True)

     # Heart_ stroke - From float to int - Coding 0 for No and 1 for Yes
    df["heart_stroke"] = (df["Heart_ stroke"].str.replace("No", "0", regex=False).str.replace("yes", "1", regex=False).astype(int))
    df.drop(columns=["Heart_ stroke"],inplace=True)

    return df

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = wrangle('/content/drive/MyDrive/Colab Notebooks/heart_disease.csv')




plt.figure(figsize=(20,10),dpi=400)
sns.heatmap(data.corr(),cmap='coolwarm'  ,annot=True,linecolor='black',linewidths=1,fmt=".3f",center=0);
```
